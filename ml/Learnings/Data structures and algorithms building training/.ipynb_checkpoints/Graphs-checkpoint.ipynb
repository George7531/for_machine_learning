# Graph:

* graph is a like a network.
* each number is called a node or vertex (so is every computer in the network)
* each number(node/vertex) has the ability to connect with other numbers(nodes/vertices). the connection is made possible through an edge.
* each edge in the network is given a weight(or cost in other words) to inform the quickest route to travel between nodes.

# unidirectional network:

* it is a one-way lane. information gets transferred from only one node to the other not the other way around!

# Bidirectional networks:

* it goes both ways. 

# Relationship between a tree,linked-list and a graph:

* linked list is a form of a binary tree and binary tree is a form of a graph which means binary tree encompasses linked lists and graphs encompasses binary trees.


# Adjacency Matrix:

* it is just like one-hot encoding i.e putting 1's in places where the component has edge with other components and 0's in places where it doesn't have any edge! 

* in a graph with weighted components the 1's will be superceded with the corresponding weights of the components!

* if it is bidirectional_(components connecting with each other) both sides from the diagonal of the matrix is symmetric(looks alike)


# Adjacency List:

* it is like a dictionary. the components(vertices) are keys the corresponding edges are values.

{

A: [B,E],

B: [A,C],

C: [B,D],

D: [C,E],

E: [A,D]

}


# Big O of Adajacency Matrix and Adjacency List:


### when adding a vertex
* Adjacency list is O(1)

* Adjacency Matrix is O(n^2)


### when removing a vertex:

* Adjacency list is O(n)

* Adjacency Matrix is O(n^2)


### when adding a new edge:

* Adjacency list is O(1)

* Adjacency Matrix is O(1)

### when removing an edge:

* Adjacency list is O(1)

* Adjacency Matrix is O(1)


# Declaring the winner between Adjacency List and Adjacency Matrix:

* the winner is (a little Drum roll ) **Adjacency List**

* Adjacency List is better from the storage perspective!

# add edge

```python
class Graph:
    # initiationg the adjacency list as a dictionary
    def __init__(self):
        self.adj_list = {}
        
        # printing the graph
    def print_graph(self):
        
        # accesses the keys of the dictionary adj_list
        for vertex in self.adj_list:
            
            # printing the keys and its corresponding values
            print(vertex,':',self.adj_list[vertex])
            
            
    def add_vertex(self,vertex): 
        
        # if the vertex is not in keys of adj_list add it as one of its key and an empty list as its
        # value
        if vertex not in self.adj_list.keys():
            self.adj_list[vertex] = []
            
            # returns True if the operation is successful else False
            return True
        
        return False
        
        
        
        
    def add_edge(self,v1,v2): # which is appending
    # only if v1 and v2 are present in the adj_list as keys connect them
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            self.adj_list[v1].append(v2)
            self.adj_list[v2].append(v1)

            # returns True if the operation succeeds 
            return True

        # returns False if the operation doesn't succeed
        return False
        
# calling the Graph class        
g = Graph() 

# setting the vertices
v1 = 1
v2 = 2

# adds vertices
g.add_vertex(v1)
g.add_vertex(v2)

# connects the edges
g.add_edge(v1,v2)

# prints the graph
g.print_graph()
```

# remove edge

```python
class Graph:
    def __init__(self):
        self.adj_list = {}
        
    def print_graph(self):
        
        for vertex in self.adj_list:
            print(vertex,':',self.adj_list[vertex])
            
            
    def add_vertex(self,vertex):
        
        if vertex not in self.adj_list.keys():
            self.adj_list[vertex] = []
            
            return True
        
        return False
    
    def add_edge(self,v1,v2):
        
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            self.adj_list[v1].append(v2)
            self.adj_list[v2].append(v1)
            
            return True
        
        return False


    def remove_edge(self,v1,v2): # which is removing

        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            
            # try the code if there's a ValueError pass and move on.
            # we do this because, 
            # sometimes the edge we mentioned to remove may not be in the list raising value error
            try:
                self.adj_list[v1].remove(v2)
                self.adj_list[v2].remove(v1)
                
            except ValueError:
                pass 

            return True

        return False
    
    
g = Graph()

v1 = 1
v2 = 2
v3 = 3

g.add_vertex(v1)
g.add_vertex(v2)
g.add_vertex(v3)

print('just the vertices as keys')
g.print_graph()
print(' ')

# edging
g.add_edge(v1,v2)
g.add_edge(v2,v3)
g.add_edge(v3,v1)

print('After edging')
g.print_graph()
print(' ')

# after removing the edge between 1 and 2
g.remove_edge(v1,v2)

print('After removing the edge between 1 and 2')
g.print_graph()
```

```python
# Test

def remove_vertex(self,vertex):
    
    # first check if the vertex you want to remove is one of the keys of self.adj_list
    if vertex in self.adj_list.keys():
        
        # if it is a key, then run through its values one by one
        
        # disconnecting the edges
        for other_vertex in self.adj_list[vertex]:
            
            # every value in the list is now seen as a key and we running through its list of values
            # and removing the specified vertex which are edges
            # from the list severing its connection with it so to speak
        
            self.adj_list[other_vertex].remove(vertex)
        
        # we are deleting the entire vertex and its values
        del self.adj_list[vertex]
        
        # returns True if the operation is succesful
        return True
    
    # returns False if it is unsuccessful
    return False
```

```python
class Graph:
    def __init__(self):
        self.adj_list = {}
        
        
    def print_graph(self):
        for vertex in self.adj_list:
            print(vertex,': ',self.adj_list[vertex])
            
    def add_vertex(self,vertex):
        if vertex not in self.adj_list.keys():
            self.adj_list[vertex] = []
            
            return True
            
        return False
    
    def add_edge(self,v1,v2):
        
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            
            self.adj_list[v1].append(v2)
            self.adj_list[v2].append(v1)
            
            return True
        
        return False
    
    def remove_edge(self,v1,v2):
        
        
        if v1 in self.adj_list.keys() and v2 in self.adj_list.keys():
            
            try:
                self.adj_list[v1].remove(v2)
                self.adj_list[v2].remove(v1)
                
            except ValueError:
                pass
            
            return True
        
        return False
    
    def remove_vertex(self,vertex):
        
        if vertex in self.adj_list.keys():
            
            for other_vertex in self.adj_list[vertex]:
                
                self.adj_list[other_vertex].remove(vertex)
                
            del self.adj_list[vertex]
            
            return True
        
        return False
                
# building the graph

print('Graph is: ')

g = Graph()

# vertices joining ceremony!
g.add_vertex('A')
g.add_vertex('B')
g.add_vertex('C')
g.add_vertex('D')

# edges(connection making)
g.add_edge('A','B')
g.add_edge('A','C')
g.add_edge('A','D')
g.add_edge('B','D')
g.add_edge('C','D')

# viewing the grandiosity of the graph made so far
g.print_graph()
print(' ')
        
# removing the vertex D from the group because it is a spy and it is trecherous.
g.remove_vertex('D')

# viewing the group after removal
print('Graph after removing vertex D: ')
g.print_graph()
```

# important points to remember:

1. adj_list is a dictionary
2. vertices are keys and edges are values.
3. add_edge is appending(in-built function of list)
4. remove_edge is removing (in-built function of list)
5. when adding or removing edges from a vertices make sure you remove them from all vertices because they are bidirectional(which means if one loses the other lose as well).
6. before removing/deleting the entire vertex make sure to disconnect all edges/connections because we want to keep the graph bidirectional, if one dies, it dies everywhere.
