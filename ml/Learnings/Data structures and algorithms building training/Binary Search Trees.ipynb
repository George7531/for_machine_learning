# Binary Tree

Binary Trees have two well connected nodes to form a complete,full, perfect trees.

Binary Trees have parents,children and siblings. if the children do not have children of their own then they are leaf nodes(its degree is one) if a node in the tree has only one edge then it is a leaf node. Two nodes from the same parent are called siblings. 

# Binary Search Tree:

Binary search Tree it does two things such as `number sorting`, `indexing according to their time of entry(first come first served)`.

it takes a node and puts it at the top. 

the second node comes in and get compared to the node at the top. if the second node is less than the top node then it goes left, if it is greater than the top node then it goes to the right. 


# big(O):

#### insert() for binary search is O(1), it is just like appending an element to a linked list.

#### lookup() and remove() for binary search tree is O(log n)

# insert


# steps:

* create a new_node.

* if the value is < the node being compared to then go left else go right.

* if the left/right to the node is empty or None then insert the new_node there otherwise repeat the above process once again. use while loop. we can't use for loop as we don't have the end. we don't know how long the tree will go. 

* if the new_node is equal to any of the other node in the tree then return False. we don't want duplicates. 

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self):
        self.root = None
        
        
    def insert(self,value):
        new_node = Node(value)
        
        if self.root is None:
            self.root = new_node
            return True
        
        temp = self.root
        
        while True:
            
            if new_node.value == temp.value:
                return False
            
            elif new_node.value < temp.value:
                if temp.left is None:
                    temp.left = new_node
                    return True
                
                else:
                    temp = temp.left
                    
            else:
                if new_node.value > temp.value:
                    if temp.right is None:
                        temp.right = new_node
                        return True
                    else:
                        temp = temp.right

tree = BinarySearchTree()

tree.insert(34)
tree.insert(87)
tree.insert(90)
tree.insert(1)
tree.insert(2)

print(tree.root.value)
print(tree.root.left.value)
print(tree.root.right.value)
            
```

# contains:

* take a value from the user and check if the value is in the BinarySearchTree

# pseudo code

step0: take the value: new_node

step1: if root == None return False

step2: temp = self.root

step3: if the temp = new_node return True. 

step4: while temp is not None, the new_node is < temp then temp = temp.left. when the temp = new_node return True

step5: while temp is not None, the new_node is > temp then temp = temp.right. when the temp = new_node return True


```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self):
        self.root = None
        
        
    def insert(self,value):
        new_node = Node(value)
        
        if self.root is None:
            self.root = new_node
            return True
        
        temp = self.root
        
        while True:
            
            if new_node.value == temp.value:
                return False
            
            elif new_node.value < temp.value:
                if temp.left is None:
                    temp.left = new_node
                    return True
                
                else:
                    temp = temp.left
                    
            else:
                if temp.right is None:
                    temp.right = new_node
                    return True
                else:
                    temp = temp.right
                        
                        
    def contains(self,value):
        new_node = Node(value)
        if self.root is None:
            return False
            
        temp = self.root
            
        while temp is not None:
                
            if new_node.value == temp.value:
                return True
                
            elif new_node.value < temp.value:
                    temp = temp.left
                    
            else:                
                temp = temp.right
                    
        return False # when temp is None return False

tree = BinarySearchTree()

tree.insert(34)
tree.insert(87)
tree.insert(90)
tree.insert(1)
tree.insert(2)

# contains check

tree.contains(88)
```
