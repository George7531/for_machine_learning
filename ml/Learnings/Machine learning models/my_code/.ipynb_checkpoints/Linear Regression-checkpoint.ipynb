# to find out how much error has the model made we have few models namely:

###  mean absolute error = (actual data - predicted data)/n 

### mean absolute error do not punish the outliers much
# $$ \text{Mean Absolute Error} = \frac{1}{n} \sum_{n}^{i} (y_i - \hat{y_i})$$

### mean squared error = ((actual data -  predicted data)**2)/2
### mean squared error punishes the outliers adequetly. The only difference is we are squaring the previous formula
# $$ \text{Mean Squared Error} = \frac{1}{n} \sum_{n}^{i} (y_i - \hat{y_i})^2 $$

### Root mean squared error:
### mean squared error method is better than mean absolute error method but root mean squared error method is even better.
### mean Squared error since it takes the square of the error values, it makes it bigger than it actually is that is compensated with taking square root of the whole which is Root mean squared error.

# $$ \text{Root Mean Squared Error} = \sqrt{ \frac{1}{n} \sum_{n}^{i} (y_i - \hat{y_i})^2} $$

### RMSE (root mean squared error of 10 rupees is good for a house but horrible for a candy)

### conclusion: mae works like standard deviation, goes to a shorter distance in covering the distance between actual values and predicted line. mse simply covers a larger distance between the predicted line and actual values, it may have actually included all the outliers within its range simply because it covers a larger distance. rmse stands between them, giving a value in the middle.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
# creating the DataFrame
df = pd.read_csv('Advertising.csv')

# defining X and y features
X = df[['TV','radio','newspaper']]
y = df['sales']

# setting the # of different polynomial degrees from 1 to the set degrees to run the polynomial_features for 
degrees = 10

# to store rmse_test and rmse_train we create empty lists
rmse_train_list = []
rmse_test_list = []


for i in range(1,degrees):

    polynomial_constructor = PolynomialFeatures(degree = i, include_bias =False)
    polynomial_features = polynomial_constructor.fit_transform(X)

    x_train, x_test, y_train, y_test = train_test_split(polynomial_features,y,test_size=0.3,random_state = 101)
    
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    
    
    # based on test data find rmse
    actual_test = y_test
    predicted_test = lr.predict(x_test)
    
    # rmse for test
    rmse_test = np.lib.scimath.sqrt(mean_squared_error(actual_test,predicted_test))
    
    
    # based on train data find rmse
    
    actual_train = y_train
    predicted_train = lr.predict(x_train)
    
    # rmse for train data
    rmse_train = np.lib.scimath.sqrt(mean_squared_error(actual_train,predicted_train))
    
    
    # appending to the empty lists (rmse_test and rmse_train)
    
    rmse_train_list.append(rmse_train)
    rmse_test_list.append(rmse_test)
    
    
    
    
    
    
    
    
    
    
```

```python
rmse_train_list

# the rmse_train_list should decrease as we go up the higher polynomial degree but here we get sudden spike (1.52) in the 
# middle which indicates that we have a problem with the training model
```

```python
rmse_test_list


# here is a clear redflag as the error increases with each degree increment. A good model does the opposite.
```

# plotting the rmse_train_list and rmse_test_list to see where it went all chaotic

```python
plt.figure(dpi=150)

x = np.linspace(1,5,5)
plt.plot(x,rmse_train_list[:5],label='RMSE_TRAIN')
plt.plot(x,rmse_test_list[:5],label='RMSE_TEST')

plt.legend()
plt.show()

# refer the linear regression page in onenote for learning the ideal,good and bad fits and overfitting & underfitting.
# in onenote everything is shown with picture and details that are easily visualizable.

# the ideal degree of polynomial in this case is 3 because that is when the error is at the lowest and show no signs of 
# sudden jump
```

# Feature scaling: 

	1. Feature scaling is normalising the metrics of each features to a common value such that each feature is relatable i.e converting movie ratings from IMDb, Rotten tomatoes, meta critic and fandango to a common scale of 1-5. 
    
	2. Feature scaling improves performance of the algorithm. 
    
	3. In some cases feature scaling is not necessary especially if ther are two unrelatable features whose metrics have no commonality or not corelatable i.e normalising or feature scaling age and height of people in the data set is not necessary.
    
	4. Linear regression, random forest model, decision trees are insensible to feature scaling. So no matter you do feature scaling on them these models won't improve in performance  whereas knn model (k nearest model is well receptive of feature scaling and its performance will have been significantly improved if feature scaling is done on right features (age, sex, occupation, credit score among others are some valid features to be scaled as they are related to each other in some ways( corelatable). 

# Two common types of features scaling:

	1. Standardization: z score normalisation $$ z = \frac{x-u}{s} $$
    
	2. Normalisation: $$ N = \frac{x-xmin}{xmax-xmin}. $$

# Important considerations:

	1.  fit only on training data set. 
    
	2. We don't want to assume prior knowledge of the test data.
    
	3. Do not scale the nominal categorical data it would make no elking sense. i.e scaling the label such as a persons' names "George", "Samuel", "Muthu", "moonju mela kuthu" won't be apprpriate. 

	
	


# cross validation

* it is used to know how our machine learning model is going to predict the future with 'unseen' data sets, as in reality.

* Training and test method may work normally but they don't account for unseen nature of the realiy so what we do with cross validation is we hold out a piece of data, that the computer is oblivious of, for final testing our model. 

* That oblivious factor is very important here because that translates to predicting the unseen data and finding how accurate the model really is.

it is a technique for splitting the training dataset and test dataset into three parts:
1. Training data set
2. validation data set (typical test set).
3. Hold out Test set. (this is that unseen data that the computer, apologies! it our model not computer, is going to be oblivious of.

* Here Training data set and test data set are used normally but we test our final model with hold out test 

# Difference between Features of the data set and Hyperparameters

* Features are labels: x features:  # of bedrooms, # of kitchens, Garden, pool. y feature: 'price of House'
* Hyperparameters are like buttons to control the given dataset; fixing the # of iterations, applying regularization, normalization, standardization, size of training and test set, choice of algorithm (linear regression, knn, decision tree, random forest among others) . 
* They are not same!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import SCORERS
```

# building the ridge regression from scratch

```python
# first DataFrame

df = pd.read_csv('Advertising.csv')

X = df.drop(columns = 'sales')
y = df['sales']

# converting the X into polynomial 
polynomial_convertor = PolynomialFeatures(degree=3,include_bias=False)
polynomial_features = polynomial_convertor.fit_transform(X)

# performing train_test_split
x_train,x_test,y_train,y_test = train_test_split(polynomial_features,y,test_size=0.3,random_state=101)

# scaling the converted X_train and x_test
scaler = StandardScaler()
scaler.fit(x_train)
scaled_x_train = scaler.transform(x_train)
scaled_x_test = scaler.transform(x_test)

# ridge regression 
ridge_model = Ridge(alpha=10)
ridge_model.fit(scaled_x_train,y_train)
predicted = ridge_model.predict(scaled_x_test)
actual = y_test

# mae,rmse
mae = mean_absolute_error(actual,predicted)
rmse = np.sqrt(mean_squared_error(actual,predicted))
```

# how do we know alpha=10 (penalty factor that makes the slope a little flatter) should be at 10 not someother value?

* that's why we do cross validation
* RidgeCV is going to perform a test on given alpha values in the tuple to find the optimal alpha for the model among the given ones.

```python
# we are supplying the RidgeCV 
# scoring is supplied to accurately pick the best alpha values based on which alpha value gives the least error
ridge_cv_model = RidgeCV(alphas=(0.1,1,10),scoring='neg_mean_absolute_error')

# then give only the training set
ridge_cv_model.fit(scaled_x_train,y_train) # it's going to leave a hold out values to test later on

# the best alpha value of the supplied value is 
print('the best alpha value is: ',ridge_cv_model.alpha_)

```

# Lasso Regression : Least Absolute Shrinkage and Selection Operator

* in LassoCV() we are setting the alpha values between a range of values for the computer to choose the best from.

* eps ( alpha_min/alpha_max)  gives a ratio. This sets the range of alphas eps = 0.08 means min alpha value = 0.08 max alpha value=1. one noteworthy thing is the max value is set to 1 by default.The eps determines the tolerance threshold. lesser the tolerance value in this case eps, higher the accuracy. but there is a tradeoff between high accuracy,computational speed and cost. 

* n_alphas is number of alphas along the regularization path. i think it is like n_alpha =100 would give 100's of alphas between min and max

* max_iter determines the maximum # of iterations allowed for the model to find the optimal values for the coefficients,and reduce the errors maximally to give better prediction model. setting a larger max_iter would give the best solution but increases computation time. 

* cv- cross validation. it determines how many parts the given data must be split into for evaluation. i.e cv=5 means split the data into 5 parts run the Lasso fit on each split. the model will be trained and evaluated 5 times. Each time it uses one fold  as  test_set and other folds as the training_set. This reduces the 'chance' of overfitting.


# when to use lasso vs ridge:

* lasso just removes the insignificant x features that are contributing much to the output. it just make these x features become zero, pruning those lateral insignificant branches. so use lasso regression when you want good precision in prediction but the trade off here is losing some of the x features' contribution to prediction.

* ridge never removes any coefficients even if they are insignificant or their contribution to the output is negligible. but ridge regression can't sometimes be as precise as lasso. 

* so when the all the features are important use ridge, when some of the features can be removed use lasso because lasso makes them become zero. 

* when you want to know which feature is important use lasso. when you think all features are important or all are correlated then use ridge


# Lasso penalty and Ridge penalty difference

# $$ \text{Lasso} = sum_{i=1}^{p} |\beta_i| <= S $$
# $$ \text{Ridge} = sum_{i=1}^{p} (\beta_i^{2}) <= S $$

```python
lasso_cv_model = LassoCV(eps=0.0001,n_alphas=1000,cv=5,max_iter=100000)
lasso_cv_model.fit(scaled_x_train,y_train)

actual = y_test
predicted = lasso_cv_model.predict(scaled_x_test)

# mae and rmse
mae = mean_absolute_error(actual,predicted)
rmse = np.sqrt(mean_squared_error(actual,predicted))
```

# Elastic Net

* Elastic Net is combining the strenth of  both ridge and lasso regression penalties to overcome some of the shortcomings of each method. this reduces the tradeoff that happens with choosing ridge or lasso exclusively.

* L1 is Lasso.

* L2 is ridge.

#  $$ \text{Elastic Net} = \text{L1 * Lasso + (1-L1)*Ridge}  $$

* L1 = 1 means complete Lasso and L1 = 0 complete Ridge.

* L1 parameter strikes the balance between using Lasso and Ridge penalty.

* The main advantage of using Elastic Net over exclusive Ridge or Lasso is it will be really useful in high dimensional data with many correlated features but we also want to prune some of the insigificance. using exclusive Lasso in this case will prune all the insignificant and set them to zero whereas an exclusive Ridge will mean keeping all the features including the insigficant ones making the model less precise to reality. 





```python
elastic_model = ElasticNetCV(l1_ratio = [0.1,0.2,0.3,0.5,0.7,0.9,0.95,0.99,1],eps=0.0001,n_alphas=100,cv=5,max_iter=1000000)

# fitting the training data
elastic_model.fit(scaled_x_train, y_train)

# defining the actual and predicted data
actual = y_test
predicted = elastic_model.predict(scaled_x_test)

# finding the mae and rmse
mae = mean_absolute_error(actual,predicted)
rmse = mean_squared_error(actual,predicted)
```

# Transforming categorical data for machine interpretation.  

* To make the model understand certain features of categories, we rank or integer encode the categories according to their effect or power. i.e spicy level: nothing:0, mild:1, hot:2, fire: 3. we are just converting categories into 'ordinal' categories. This is useful when there is relationship between the features(x). this doesnot increase the number of features or number of coefficients. 

* when there is no relationship between the features of x, we use 'one hot encoding'. Assinging Boolean(yes/no) or some random numbers to the categories might still serve the purpose of making the data interpretable and operable by the algorithm with the benefit of not adding any order to it. Be cautious thought with Assigning random numbers as it may interpreted with biasedness if not beforehand the interpreter is given the knowledge about the randomness of the number assignment. This increases the number of coefficients/ number of features depending upon the number of categories. Domain knowledge/ expertise in the field or information from the experts might help us reduce the total number of features by not micromanaging or nano-cutting. 


# outliers

* Iqr method: high = q3+1.5*(IQR), low = q1-1.5*(IQR)
* standard deviation method: 4 std above
* determining the outliers with domain experts or expertise



```python
# demonstrating outliers

# defining mean,std and n for creating a normal distributed sample_ages
mean = 50
std = 13
n = 100

# setting the seed so that the result doesn't change at every flick of the button.
np.random.seed(42)

# creating normally distributed sample_ages
sample_ages = np.random.normal(loc=mean,scale=std,size=n)
sample_ages = np.round(sample_ages,decimals=0)


# plotting that in seaborn
plt.figure(dpi=100)
sns.histplot(data = sample_ages,bins=20)

plt.show()

# create a box plot to see if we have outliers
sns.boxplot(data=sample_ages,orient='h')

plt.show()
```
