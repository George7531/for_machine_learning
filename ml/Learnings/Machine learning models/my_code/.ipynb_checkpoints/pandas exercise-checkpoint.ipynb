Now let's deal with missing data.

```python
# Import the car sales DataFrame with missing data ("../data/car-sales-missing-data.csv")
df_missing = pd.read_csv('car-sales-missing-data.csv')

# Check out the new DataFrame
df_missing
```

Notice the missing values are represented as `NaN` in pandas DataFrames.

Let's try fill them.

```python
# Fill the Odometer column missing values with the mean of the column inplace

df_missing['Odometer'] = df_missing['Odometer'].fillna(np.round(df_missing['Odometer'].mean(),2))
df_missing
```

```python
# View the car sales missing DataFrame and verify the changes
df_missing
```

```python
# Remove the rest of the missing data inplace
df = df_missing.dropna()
df
```

```python
# Verify the missing values are removed by viewing the DataFrame
df
```

We'll now start to add columns to our DataFrame.

```python
# Create a "Seats" column where every row has a value of 5
seats = 5

df['Seats'] = seats
df
```

```python
# Create a column called "Engine Size" with random values between 1.3 and 4.5
# Remember: If you're doing it from a Python list, the list has to be the same length
# as the DataFrame

mini = 1.3
maxi = 4.5
engine_sizes = np.random.rand(len(df))*maxi + mini

df['Engine Size'] = engine_sizes

df
```

```python
df['Price'] = df['Price'].str.replace('[$,.]','').astype(int)
```

```python
# Create a column which represents the price of a car per kilometer
# Then view the DataFrame


df['Price/KM'] = df['Price']/df['Odometer']
df
```

```python
# Remove the last column you added using .drop()
df = df.drop('Price/KM',axis=1)

df
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the the shuffled DataFrame to a new variable
df_shuffled = df.sample(frac=1)
df_shuffled
```

Notice how the index numbers get moved around. The [`sample()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html) function is a great way to get random samples from your DataFrame. It's also another great way to shuffle the rows by setting `frac=1`.

```python
# Reset the indexes of the shuffled DataFrame
df_shuffled.reset_index(drop=True)
```

Notice the index numbers have been changed to have order (start from 0).

```python
# Change the Odometer values from kilometers to miles using a Lambda function
# Then view the DataFrame
df['Odometer'] = np.round(df['Odometer'].apply(lambda x: x/1.609),2)

df
```
