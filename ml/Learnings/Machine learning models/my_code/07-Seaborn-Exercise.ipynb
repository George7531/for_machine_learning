# normal pointer (seperate memory after update,taking more space)

```python
a=2
b=a

# before update

print('Before update')
print('a = ', a)
print('b = ', b)

# after update
b=3

print('\nAfter value is updated')
print('a =',a)
print('b =',b)
```

# dictionart pointer. after updated not only the dictionary that underwent change/update that had changed but the dictionary it had associated with before change changes too. Fascinating!

```python
dict1= {'George': 97}
dict2= dict1

# before update

print('Before update')
print('dict1 = ', dict1)
print('dict2 = ', dict2)

# after update
dict2['George']= 98

print('\nAfter value is updated')
print('dict1 =',dict1)
print('dict2 =',dict2)
```
