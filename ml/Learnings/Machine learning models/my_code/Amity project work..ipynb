### Creating Cohorts based on Tenure

**Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.**

**Treating each unique tenure group as a cohort, calculating the Churn rate (percentage that had Yes Churn) per cohort. For example, the cohort that has had a tenure of 1 month should have a Churn rate of 61.99%. we should have cohorts 1-72 months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as we are less likely to stop service the longer you've had it.**

```python
df.head().T
```

step 1: get dummies for churn 

step 2: get tenure_cohorts unique values

step 3: check if the tenure is in tenure_cohorts and churn is 'Yes'. 

```python



tenure_cohorts = df['tenure'].sort_values().unique()

churned = pd.get_dummies(df['Churn'],drop_first=True)

cr = []

for i in range(len(tenure_cohorts)):
    yes = sum((df['tenure'] == tenure_cohorts[i]) & (df['Churn'] == 'Yes'))
    no = sum((df['tenure'] == tenure_cohorts[i]) & (df['Churn'] == 'No'))
    cr.append(100*yes/(yes+no))
    
    

d = {'Tenure':list(tenure_cohorts),'Churn_percentage':cr}
churn_rate = pd.DataFrame(data=d)
churn_rate = churn_rate.set_index('Tenure')
churn_rate
```

```python
churned.sum()
```

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
yes_churn
```

```python
churn_rate = 100* yes_churn/(no_churn + yes_churn)

churn_rate.T['customerID']
```

**Now that we have Churn Rate per tenure group 1-72 months, creating a plot showing churn rate per months of tenure.**

```python
plt.figure(dpi=150,figsize=(12,4))
plt.plot(churn_rate.index,churn_rate['Churn_percentage'])
plt.xlabel('Tenure')
plt.ylabel('Churn_Percentage')



plt.show()
```

### Broader Cohort Groups
**Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '12-24 Months'
   * '24-48 Months'
   * 'Over 48 Months'    

```python
df.head(5).T
```

```python
def tenure_cohort(tenure):

    if tenure <=12:
        return '0-12 Months'

    elif (tenure >12 ) & (tenure <=24):
        return '12-24 Months'

    elif (tenure >24 ) & (tenure <= 48):
        return '24-48 Months'


    elif tenure > 48:
        return 'Over 48 Months'

df['Tenure Cohort'] = df['tenure'].apply(tenure_cohort)      
```
