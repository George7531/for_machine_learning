# Hierarchical clustering

* This technique is borrowed from biology

# Agglomerative clustering:

* join the seperate clusters into one

# Divisive clustering:

* spread out the joint clusters into many.( divide the united). 

# three hot mechanisms of Hierarchical clustering:

* similarity metric.

* dendrogram.

* linkage matrix.

# similarity metric:

1. see how close the points are to the central mean using eucledian distance.


2. use min max scaling using normalization technique (data - min(data))/(max(data)-min(data))


# dendrogram:

1. shows all potential clusters for the data set.

2. computationally expense(time sucking algorithm).

3. can be very useful if the dataset is small.

# linkage matrix:

Linkage answers these two questions:

1. how do we measure the distance of a point to the entire cluster it 

2. how do we measure the distance of one cluster to another cluster.

# let us do the agglomerative clustering

* here we have to choose the n_clusters ourselves. A small technical intuition as to how to select the n_clusters is observe the cluster map and choose the most reasonable number. becuase of this implied subjectivity attached to the selection of n_clusters the agglomerate clustering is less preferred especially in cases where cluster map visualization is not computationally time-comsuming and energy draining.

# setting the distance_threshold

read the error code above. either distance_threshold or n_clusters needs to be set and the other has to be None

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=None,distance_threshold=0)

cluster_label = model.fit_predict(scaled_df)

cluster_label
```

```python
model.children_
```

```python
from scipy.cluster import hierarchy

linkage_matrix = hierarchy.linkage(model.children_)
linkage_matrix
```

# conclusion:

linkage_matrix conveys what rows are together and how long the distance span and how many points are underneath it. 


### i.e in the top row we see 67, 161, 1.41421356, and 2 what it means is 67 and 161 are connected by a distance of 1.41421356 and this small cluster has 2 points connected to it.

```python

plt.figure(figsize=(20,10))
dendrogram = hierarchy.dendrogram(linkage_matrix)

plt.xticks(rotation=90)

plt.show()


# how would you make use of this indecipherable dendrogram?
```

# making sense of the dendrogram to choose the ideal n_cluster

```python
plt.figure(figsize=(20,10))
dendrogram = hierarchy.dendrogram(linkage_matrix,truncate_mode='lastp',p=48)
```

# DBSCAN PROCEDURE:

1. pick a random point that is not yet assigned.



2. Determine the point type(core, border, outlier)



3. once the core point is identified add all the directly reachable points to the same cluster as the core is in. the distance is determined by the epsilon.




4. Repeat the procedure/algorithm until all points have their own individual cores and clusters. the number of points to be included in a cluster is predetermined by minimum number of points hyperparameter.
