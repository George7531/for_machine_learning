# inner merge

# inner merge is intersection: $$ A \cap B $$



```python
registration=pd.DataFrame({"registration_id":[1,2,3,4],"names":["George","dick","Andrew","lazydick"]})
logins=pd.DataFrame({"login_id":[1,2,3,4],"names":["moth","jetix","George","dick"]})
```

```python
registration
```

```python
logins
```

```python
result=pd.merge(left=registration,right=logins,how="inner",on="names")
result
```

# left and right merge

# left merge A-B
# $$ (A + A \cap B) - B $$

```python
registration
```

```python
logins
```

```python
result=pd.merge(left=registration,right=logins,how="left",on="names")
result
```

# right merge B-A

# $$ (B + A \cap B) - A $$

```python
result=pd.merge(left=registration,right=logins,how="right",on="names")
result
```

# outer merge

# outer merge is Union: $$ A\cup B $$

```python
registration
```

```python
logins
```

```python
result=pd.merge(left=registration,right=logins,how="outer",on="names")
result
```

# dealing with pivot tables

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Sales_Funnel_CRM.csv")
```

```python
df
```

```python
licenses=df[["Company","Product","Licenses"]]

licenses
```

```python
help(pd.pivot)
```

```python
pd.pivot(data=licenses,index="Company",columns="Product",values="Licenses")
```

# lets see how pivot_table works

```python
pd.pivot_table(data=df,index="Company",aggfunc="sum")
```

# another way to do the same

```python
test=df.groupby(["Company", "Account Number"]).sum()
test
```

# usefulness of pivot_table is in being able to mention exactly the values that you want the sum operation to be performed on

```python
# please do be mindful that the in 'sum' 's' is small

pd.pivot_table(data=df,index="Company",aggfunc= "sum",values=["Licenses","Sale Price"])
```

```python
pd.pivot_table(data=df,index=["Account Manager","Contact","Company"],values="Sale Price",aggfunc=["sum","mean"])
```

```python
# we can also use np.sum instead of sum
```

```python
pd.pivot_table(data=df,index=["Account Manager","Contact","Company"],values="Sale Price",aggfunc=[np.sum,np.mean])
```

# margins=True gives grand total at the bottom

```python
pd.pivot_table(data=df,index=["Account Manager","Contact","Company","Product"],values="Sale Price",aggfunc=[np.sum],\
               margins=True)
```
