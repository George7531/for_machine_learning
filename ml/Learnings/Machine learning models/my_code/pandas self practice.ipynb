```python
#sales data from different places for q1 and q2

sales_q1={"Japan":101,"Brazil":56,"Mexico":20,"USA":300,"India":110,"China":460}
sales_q2={"Japan":150,"Greece":40,"Mexico":35,"USA":450,"India":215,"China":420}
```

```python
#maki8ng  panda Series for those dictionaries

q1=pd.Series(sales_q1)
q2=pd.Series(sales_q2)

```

```python
#summing them up
q1+q2

#note the NaN section python knows to do the summing operation pair wise but since Brazil and Greece are not present in both
#dictionaries the convinient way for pandas is to fill for the missing data with NaN
```

```python
# doing the sum with fill_value to get the right value instead of NaN

q1.add(other=q2,fill_value=0) 

#it just fill ups for the missing value in the corresponding quarter with 0 and adds the value up using the value from 
# the other quarter
```

# create your own functions in pandas using apply()

we can create our own funtion apply that to DataFrame using apply function


```python
# we can create our own funtion apply that to DataFrame using apply function

#lets grab last 4 digit of the credit card number 

# creating functions to grab last 4 digits of number that is first converted to str
'''only a string object can be indexed and called so make sure to convert the num first to a string and index it and convert the 
whole into an integer '''


def last_four(num):
    return int(str(num)[-4:])


df["CC Number"].apply(last_four) #just mention the function name do not call it like this : last_num()




```

```python
df["last_four_cc"]=df["CC Number"].apply(last_four) #just mention the function name do not call it like this : last_num()
df
```

```python
# finding mean of total bill

df["total_bill"].mean()
```

```python
# $ for '<10' and $$ for '>10 and < 30' and $$$ for '>30'

def weightage(price):
    if price < 10:
        return '$'
    elif price >= 10 and price <= 30:
        return '$$'
    else:
        return '$$$'



df["weightage"]=df["total_bill"].apply(weightage)

df
```

```python
##filter data that has $$$ in weightage column and size 2

boolean=(df["weightage"] == "$$$" ) & (df["size"] == 2)
df[boolean]
```

```python


#see if the tip is generous or not tip is > 25% of the total bill or not

def generous(total_bill,tip):
    if (tip/total_bill) > 0.25:
        return "Yes"
    else:
        return "No"
    
```

# timeit function to see how long a code takes to run in other words how efficient the line of code is

```python
import timeit
```

```python
# setup code

setup= '''
import pandas as pd
import numpy  as np
df=pd.read_csv("tips.csv")
def generous(total_bill,tip):
    if (tip/total_bill) > 0.25:
        return "Yes"
    else:
        return "No"
'''
```

```python
# code snippets whose efficiency needs to be measured

stmt1='''
df["Generous"]=df[["total_bill","tip"]].apply(lambda df: generous(df["total_bill"],df["tip"]),axis=1)
'''

stmt2='''
df["Generous"]=np.vectorize(generous)(df['total_bill'],df['tip'])
'''
```

```python
#timeit function

timeit.timeit(stmt=stmt1,setup=setup,number=1000)
```

```python
timeit.timeit(stmt=stmt2,setup=setup,number=1000)
```
