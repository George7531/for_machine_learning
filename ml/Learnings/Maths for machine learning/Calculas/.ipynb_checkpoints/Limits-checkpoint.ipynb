# zeno's paradox:

```python
# exercise 2: Approximate the limit via zeno's paradox.

# trying to create the zeno's half distances for some iterations

final_distance = 1

domain = [final_distance-1,final_distance+1]


x_left = (final_distance-domain[0])/2
x_right = (final_distance+domain[1])/2


n = 10
left = np.zeros((n,2))
right = np.zeros((n,2))

def y(x):
    return np.cos(x**2)**2 + np.pi

for i in range(10):
    
    left[i,0] = x_left
    right[i,0] = x_right
    
    
    

    x_left = x_left + (final_distance-x_left)/2
    y_left = y(x_left)
    
    x_right = (final_distance+x_right)/2
    y_right = y(x_right)
    
    
    left[i,1] = y_left
    right[i,1] = y_right
    

    
    
    
print('Limit from the Left')
print(left)
print('\n')
print('Limit from the Right')
print(right)
print('\n')
print('Function at value x=1')
print(y(x=1))
    

    
 
    
    
    
    
```

# find limits of the functions from both left and right sides

* limit for a function as x approaches a point exist only if the x approaches the same point from both left and right side.

```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
sym.init_printing()
import matplotlib.pyplot as plt
```

```python
# viewing the graphical representation of a function (fx)
x = sym.symbols('x')
fx = sym.Abs(x-2)/(x-2)

p = sym.plot(fx,show=False)

p.show()


# conclusion:

'''limit does not exist for this equation as approaches 2 because as x approaches 2 from left side (possibly with a value 
of 1.999999999)is equal to -1 and as the x approaches 2 from the right side (possibly with a value of 2.000000001) 
is equal to 1 since the left side and the right side do not match the limit for this function as x approaches 2 doesn't 
exist '''
```

```python
# exercise 1: implementing limits in numpy

# defining x boundaries
xx = np.linspace(0.1,2*np.pi,213)

# defining the function
def fx(x):
    return np.cos(x*np.pi) + np.log(x)**2

# find limit for the function as x approaches 3

# zeno's paradox algorithm to find the x as it approaches the limit
limit = 3
domain = np.array([limit-1,limit+1])

# number of iterations
n = 40

# defining x and y's to capture the points as we get closer to the limit
'''the 2 columns indicate the x's points as it approaches from the left and right'''
x = np.zeros((n,2))
y = np.zeros((n,2))

# running the loop to find points closer to the limit as we approach from left and right of the limit
for i in range(n):
    x[i,:] = domain
    y[i,:] = fx(domain)
    
    # update the domain
    domain = (domain + limit)/2
    
from_left = np.vstack((x[:,0],y[:,0])).T
from_right = np.vstack((x[:,1],y[:,1])).T


# plotting the function
plt.figure(dpi=150)
plt.plot(xx,fx(xx),'k-',label='f(x)')

# plotting the points from left and right
plt.plot(from_left[:,0],from_left[:,1],'s',label='From the Left')
plt.plot(from_right[:,0],from_right[:,1],'o',label='From the Right')

plt.xlabel('x',fontsize=15)
plt.ylabel('fx',fontsize=15)
plt.grid()
plt.title('$ \\lim_{x \\rightarrow %s}$ f(x) is %s '%(limit,fx(3)))
plt.legend()


plt.show()

# conclusion:
"""approaching the same point from Left and Right so limit do exist for this equation with the given constraints"""
```

```python
# exercise 2: finding the limit of a function in sympy

x = sym.symbols('x')
fx = sym.cos(x*sym.pi) + sym.log(x)**2

# finding the limits
limit = 3
fromLeft = sym.limit(fx,x,limit,dir='-')
fromRight = sym.limit(fx,x,limit,dir='+')
limit_at_3 = sym.limit(fx,x,limit,dir='+-')

display(Math('\\text{Limit from Left of %s is: } %s'%(limit,sym.latex(fromLeft))))
display(Math('\\text{Limit from Right of %s is: } %s'%(limit,sym.latex(fromRight))))
display(Math('\\text{Limit at %s is: } %s'%(limit,sym.latex(limit_at_3))))

# conclusion:
'''since limit of the function from left,right and at the point are all same we conclude that the limit for the function
for the given constraint exist! '''
```

# properties of limits:

### factorable: lim(constant * f) = constant* lim(f)

### additive: lim(f+g) = lim(f) + lim(g)

### multiplicative: lim(f*g) = lim(f) * lim(g)

### divisive: lim(f/g) = lim(f) / lim(g)


# limits are factorable

# prove lim(constant*f) = constant * lim(f)

# $$ f(x) = \frac{x^3}{3} + 100 \sqrt{|x|} $$

# prove now:
you can insert any number for c:
# $$ \lim_{x \rightarrow 5} (cf(x)) = c * (\lim_{x \rightarrow 5} f(x)) $$

```python
c = 0.2

x = sym.symbols('x')

fx = (x**3/3) + 100* sym.sqrt(sym.Abs(x))

limit = 5

# sym.N is for numpification.
lhs = sym.N(sym.limit(c*fx,x,limit,dir='+-'))
rhs = sym.N(c*sym.limit(fx,x,limit,dir='+-'))


# displaying the materials:
display(Math('lhs: %s'%(sym.latex(lhs))))
display(Math('rhs: %s'%(sym.latex(rhs))))

print(lhs==rhs)
```

# prove limits are additive:

# $$ \lim_{x \rightarrow \pi} (f+g) =  \lim_{x \rightarrow \pi} (f) +  \lim_{x \rightarrow \pi} (g) $$

```python
# defining the symbol in sympy
x = sym.symbols('x')

# defining the functions
f = sym.log(x) + x**2
g = sym.exp(-x) + x**3

# limit as x approaches
limit = sym.pi


lhs = sym.limit(f+g,x,limit,dir='+-')
rhs = sym.limit(f,x,limit,dir='+-') + sym.limit(g,x,limit,dir='+-')

# display the result:
display(Math('\\text{lhs: } \\quad %s'%(sym.latex(lhs))))
display(Math('\\text{rhs: } \\quad %s'%(sym.latex(rhs))))

# proof of equality:
print(lhs == rhs)

# conclusion:
'''Hence proved! '''
```

# prove limits are multiplicative

# 1. $$ \lim_{x \rightarrow \pi} (f*g) = \lim_{x \rightarrow \pi} (f) * \lim_{x \rightarrow \pi} (g) $$

# 2.  $$ \lim_{x \rightarrow \pi} (f^{3}) = \lim_{x \rightarrow \pi} (f) * \lim_{x \rightarrow \pi} (f) * \lim_{x \rightarrow \pi} (f) $$

```python
# 1.

# defining the symbol in sympy
x = sym.symbols('x')

# defining the functions
f = sym.log(x) + x**2
g = sym.exp(-x) + x**3

# limit as x approaches
limit = sym.pi


lhs = sym.limit(f*g,x,limit,dir='+-')
rhs = sym.limit(f,x,limit,dir='+-') * sym.limit(g,x,limit,dir='+-')

# display the result:
display(Math('\\text{lhs: } \\quad %s'%(sym.latex(lhs))))
display(Math('\\text{rhs: } \\quad %s'%(sym.latex(rhs))))

# proof of equality:
print(lhs == rhs)


# conclusion:
'''Hence Proved!'''
```

```python
# 2.

# defining the symbol in sympy
x = sym.symbols('x')

# defining the functions
f = sym.log(x) + x**2

# limit as x approaches
limit = sym.pi


lhs = sym.limit(f**3,x,limit,dir='+-')
rhs = sym.limit(f,x,limit,dir='+-') * sym.limit(f,x,limit,dir='+-') * sym.limit(f,x,limit,dir='+-')

# display the result:
display(Math('\\text{lhs: } \\quad %s'%(sym.latex(lhs))))
display(Math('\\text{rhs: } \\quad %s'%(sym.latex(rhs))))

# proof of equality:
print(lhs == rhs)

# conclusion:
'''Hence Proved!'''
```

# prove Limits are Divisive:

# 1. $$ \lim_{x \rightarrow \pi} (f/g) = \frac{\lim_{x \rightarrow \pi} (f) }{\lim_{x \rightarrow \pi} (g) } $$

# 2. $$ \lim_{x \rightarrow 0} (g/h) = \frac{\lim_{x \rightarrow 0} (g) }{\lim_{x \rightarrow 0} (h) } $$

```python
# 1.

# defining the symbol in sympy
x = sym.symbols('x')

# defining the functions
f = sym.log(x) + x**2
g = sym.exp(-x) + x**3

# limit as x approaches
limit = sym.pi


lhs = sym.limit(f/g,x,limit,dir='+-')
rhs = sym.limit(f,x,limit,dir='+-') / sym.limit(g,x,limit,dir='+-')

# display the result:
display(Math('\\text{lhs: } \\quad %s'%(sym.latex(lhs))))
display(Math('\\text{rhs: } \\quad %s'%(sym.latex(rhs))))

# proof of equality:
proof = lhs == rhs

print(proof)


# conclusion:
if proof:
    
    print('Hence Proved!')

```

```python
# 2.

# defining the symbol in sympy
x = sym.symbols('x')

# defining the functions
g = sym.exp(-x) + x**3
h = x**3 + x**2 + x

# limit as x approaches
limit = 0


lhs = sym.limit(g/h,x,limit,dir='+-')
rhs = sym.limit(g,x,limit,dir='+-') / sym.limit(h,x,limit,dir='+-')

# display the result:
display(Math('\\text{lhs: } \\quad %s'%(sym.latex(lhs))))
display(Math('\\text{rhs: } \\quad %s'%(sym.latex(rhs))))

# proof of equality:
proof = lhs == rhs

print(proof)


# conclusion:
if proof:
    
    print('Hence Proved!')

```

# continuous:
* A function f(x) is continuous if the following conditions are met:


# 1. $$ \lim_{x \rightarrow a } f(x) \text{ exists} $$
# 2. $$ \lim_{x \rightarrow a } f(x) = f(a)  $$



* a is any constant
* the simple solution to check whether the function is continuous is to plot the function.

# discontinuous:

* if there is a gap in the function f(x) when plotted then it is discontinuous.
* there are 3 types of discontinuity namely, jump discontinuity, infinite discontinuity, removable discontinuity.

\end{pmatrix} $$

# Limits of Trignometric Functions;

# Two important Trig limits to know:

# 1. $$ \lim_{\phi \rightarrow 0} \{ \frac{\cos{\phi} - 1}{\phi} \} = 0 $$

# 2. $$ \lim_{\phi \rightarrow 0} \{ \frac{\sin{\phi} }{\phi} \} = 1 $$

# explanation for 2:

* draw a unit circle with radius = 1 unit.

* draw a line from the origin to touch a point on the circle.

* phi is the angle 

* the hypotenuse(from the point on the circle to the x cooridinate) is the sin

* phi the angle is always going to larger than the hypotenuse the sine.

* so the ratio of the hypotenuse (sine) and angle (phi) is always going to fall between 0 and 1.

* but as the angle gets smaller and smaller in size the ratio of hypotenuse(sine) and the angle(phi) gets to 1.

* that's how we got 1.


# explanation for 1:

* $$ \lim_{x \rightarrow 0} \{ \frac{\cos x - 1}{x } \} = 0 $$
* $$ \downarrow $$
* $$ =\frac{\cos x - 1}{x } * \frac{\cos x + 1}{\cos x + 1} $$
* $$ \downarrow $$
* $$ =\frac{\cos^{2} x -1 }{x(\cos x + 1) } \rightarrow eq1 $$
* $$ \downarrow $$
* $$ =\sin^2 x + \cos^2 x = 1 \rightarrow \cos^2 x -1 = - \sin^2 x \rightarrow eq2 $$
* $$ \text{eq2 in eq1} $$
* $$ \downarrow $$
* $$ =\frac{-\sin^2 x }{x(cos^2 x + 1)} $$
* $$ \downarrow $$
* $$ =\lim_{x \rightarrow 0} \{\frac{\sin x }{x} *  \frac{-\sin x}{\cos x + 1 } \}$$
* $$ \downarrow $$
* $$ =\lim_{x \rightarrow 0} \{ 1 * \frac{-sin x}{\cos x + 1} \} $$
* $$ \downarrow $$
* $$ =\frac{0}{1 + 1} =\frac{0}{2}  = 0 $$

# squeeze theorem:

* it is just another approximation tool that finds the value of funtion fx when x approaches a value that value when directly plugged into the fx might go to infinity and cannot find the limit for, but other two functions gx,hx might come to fx's rescue by offering themselves to find the limit as x approaches that value in them(in gx and hx as x approaches the same value).
* but the condition is gx and hx must come close or touch fx at that same value that x approaches that we find the limit for.
* one useful advice to see whether all fx,hx and gx touch eachother at the same value as x apporoaches is by plotting them. looking at their geormetry helps.

# architechture:
# $$ \lim_{x \rightarrow a} g(x) \le \lim_{x \rightarrow a} f(x) \le \lim_{x \rightarrow a} h(x) $$

# application of squeezing:
* find the limit of a function that is otherwise impossible to compute.
* find the derivative of a function that is otherwise impossible to compute.

# solve: $$ \lim_{x \rightarrow 0} \left[x^2 \sin \frac{1}{x} \right] $$

```python
x = np.linspace(-0.5,0.5,1000)
fx = x**2 * np.sin(1/x)
hx = x**2
gx = -x**2

# plotting:
plt.figure(dpi=150)
plt.plot(x,fx,label='fx')
plt.plot(x,hx,label='hx')
plt.plot(x,gx,label='gx')


plt.xlim([-0.5,0.5])
plt.ylim([-0.2,0.2])


plt.legend()
plt.show()

# conclusion:
'''even though in the function fx as x approaches 0 the fx becomes infinitely oscillating the gx and the hx squeeze fx
to give a value of 0 as x approaches 0 in all fx,gx,hx this is what squeeze theorem is about!'''
```

# goal:

prove: 

# $$ \frac{\tan(x)} {x} = \frac{\sin(x)}{x} \frac{1}{cos(x)} $$

* 1. implement the function in sympy
* 2. plot using matplotlib.
* 3. compute the function values as x=0
* 4. compute and print the limits in sympy as x approaches 0.


```python
# 1. implement the function in sympy

# define the variables to be used inside the function.
x = sym.symbols('x')

# implementing the function in sympy
lhs = sym.tan(x)/x
f1 = sym.sin(x)/x
f2 = 1/sym.cos(x)
rhs = f1 * f2

sym.Eq(lhs,rhs)
```

```python
# 2.plot using matplotlib
xx = np.linspace(-3,3,1000)

def numpification(fn):
    return sym.lambdify(x,fn,'numpy')

lhs_numpy,f1_numpy,f2_numpy,rhs_numpy = map(numpification,(lhs,f1,f2,rhs))

# plotting
plt.figure(dpi=150)
plt.plot(xx,lhs_numpy(xx),label='LHS = $ \\frac{\\tan(x)}{x} $ ')
plt.plot(xx,f1_numpy(xx),label='f1 = $ \\frac{\\sin(x)}{x} $')
plt.plot(xx,f2_numpy(xx),label='f2 = $ \\frac{1}{\\cos(x)} $')
plt.plot(xx[::18],rhs_numpy(xx)[::18],'ro',label='RHS = $ \\frac{\\sin(x)}{x \\cos(x)} $')

# zooming in a bit
plt.ylim([-20,20])


# decorating a bit
plt.grid()
plt.legend()

plt.show()

# conclusion:
'''conclusion is lhs = rhs as the fact has been proven by red dots hovering just over the blue line'''
```

# singularity types:

**Removable Singularity**: At a removable singularity, the function is undefined at that point, but it can be modified or "remedied" by defining or redefining the function at that point to make it continuous and well-behaved.

**Pole**: A pole is a type of singularity where the function approaches either positive or negative infinity as you get closer to the singular point. The function has a vertical asymptote at the pole, and its magnitude increases without bound as you approach the singularity.

**Essential Singularity**: An essential singularity is a point where the function's behavior is extremely complicated and does not fit into the categories of removable singularities or poles. At an essential singularity, the function exhibits highly oscillatory or chaotic behavior.

**Branch Point**: In complex analysis, branch points are points where the function becomes multi-valued or exhibits different values as you go around the point in different ways. These points are often associated with functions that have complex values.

note: The presence of a singularity often indicates a challenging aspect of the function, and it requires careful analysis to understand and work with the function in the vicinity of that point. Depending on the nature of the singularity, it may be possible to apply techniques such as Laurent series expansion, residue theorem, or contour integration to analyze and handle the singularity appropriately.

# exercise 1:

# $$ \lim_{x \rightarrow a} \frac{c\left(x^{2} -(a+b)x + ab \right)}{\left(x^{2} + (b-a)x -ab \right)} $$

$$ a,b \in \{-4,-3,-2,-1,0,1,2,3,4\} $$

$$ c \in \{-1,1\} $$

```python
x,a,b,c = sym.symbols('x,a,b,c')
fx = c*(x**2 -(a+b)*x +a*b)/(x**2 + (b-a)*x -a*b)


# setting a,b,c
sub_a = (-4,-3,-2,-1,0,1,2,3,4)
sub_b = (-4,-3,-2,-1,0,1,2,3,4)
sub_c = (-1,1)

# choosing the index to see value for:
index_for_a = 6
index_for_b = -1
index_for_c = 1

# substituting
print('fx after substitution of a,b,c in it')
result = fx.subs({a:sub_a[index_for_a],b:sub_b[index_for_b],c:sub_c[index_for_c]})
display(result)

# plotting
p = sym.plot(result,show=False)
p.ylim = [-10,10]
p.show()

# limit of result
limit = sub_a[index_for_a]

print('Limit as x>{} from left: {}'.format(limit,sym.limit(result,x,limit,dir='-')))
print('Limit as x>{} from right: {}'.format(limit,sym.limit(result,x,limit,dir='+')))
print('Limit as x>{} Two Sided:  {}'.format(limit,sym.limit(result,x,limit,dir='+-')))
print('Function at value {} is: {}'.format(limit,result.subs({x:limit})))
```

# exercise 2: square roots and conjugates.

**side note: conjugate is changing the sign of a imaginary part of the complex number i.e from (a+bi) -> (a-bi.)**

# solve for: 
# $$ \lim_{x \rightarrow -a} \frac{\left(\sqrt{x+b} + c \right)}{\left(x + a \right)} $$

$$ a \in \{-4,-3,-2,-1,0,1,2,3,4\} $$

$$ b \in a + \{0,1,2,3,4 \} $$

$$ c = -\sqrt{b-a} $$

```python
# setting a,b,c
a = np.random.choice((-4,-3,-2,-1,0,1,2,3,4))
b = a + np.random.choice((0,1,2,3,4))
c = np.round(-np.sqrt(b-a),4)
print('values of a,b,c: {}'.format((a,b,c)))
print(' ')

# defining the substituable symbol which is x
x = sym.symbols('x')

# writing the function
print('Function is: ')
fx = (sym.sqrt(x+b) +c)/(x+a)
display(fx)

# plotting
p = sym.plot(fx,show=False)
p.show()

# limit of result
limit = -a

print('Limit as x>{} from left : {}'.format(limit,sym.limit(fx,x,limit,dir='-')))
print('Limit as x>{} from right: {}'.format(limit,sym.limit(fx,x,limit,dir='+')))
print('Limit as x>{} Two Sided :  {}'.format(limit,sym.limit(fx,x,limit,dir='+-')))
print('Function at value {} is : {}'.format(limit,fx.subs({x:limit})))
```
