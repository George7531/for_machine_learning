$$ \text{derivative for k(x) is: }  \lim_{\Delta x \rightarrow 0} \left[\frac{k(x+\Delta x) -k(x)}{\Delta x} \right] $$

$$ \downarrow $$

# prove: Differentiability also implies Continuity:

# notes:
* Axiom is if a differentiability exist for point x then the point x is also continuous.
* that means a discrete point is non-differentiable.



Your professor is correct. The non-symmetry between differentiability and continuity is an interesting aspect of real analysis, and it arises from the different requirements and definitions of these concepts.

Let's delve into why differentiability implies continuity, while the reverse is not always true:

### Differentiability Implies Continuity:

The definition of differentiability involves the existence of a certain limit:

$$ \left[ \lim_{{h \to 0}} \frac{{f(c + h) - f(c)}}{h} \right] $$

This limit must exist for \( f(x) \) to be considered differentiable at a point \( c \). When this limit exists, it means that as \( h \) approaches 0, the difference quotient approaches a well-defined value, which is the derivative of \( f \) at \( c \).

The crucial point here is that if the limit exists, it implies that the function is "smooth" enough near the point \( c \). As \( h \) gets arbitrarily close to 0, the function values \( f(c + h) \) and \( f(c) \) must approach each other, leading to continuity at \( c \).

### Continuity Does Not Necessarily Imply Differentiability:

Continuity, on the other hand, only requires that the function values are close to each other as \( x \) gets close to a certain point. This closeness is determined by the definition of continuity:

$$ \left[ \lim_{{x \to c}} f(x) = f(c) \right] $$

This does not guarantee that the rate of change, i.e., the derivative, exists at that point. The function might have abrupt changes, corners, or vertical tangents, which make the derivative undefined.

### Examples:

1. **Continuous but Not Differentiable:**
   - The absolute value function \( f(x) = |x| \) is continuous everywhere but not differentiable at \( x = 0 \).

2. **Nowhere Differentiable:**
   - There are functions, like the Weierstrass function, that are continuous everywhere but differentiable nowhere. These functions exhibit intricate oscillations and do not have a well-defined derivative at any point.

In summary, differentiability implies a certain level of smoothness and local linearity, leading to continuity. However, continuity alone does not provide enough information about the local behavior of the function to guarantee differentiability. This non-symmetry is a fascinating aspect of the mathematical properties of functions.

# exercise 1 confirm the product rule

# $$ (fg)^{'} == f^{'}g + g^{'}f $$

```python
# exercise 1 confirm the product rule

x = sym.symbols('x')
fx = x**2
gx = sym.cos(x)

# prove: (fg)^{'} = f^{'}g + g^{'}f
lhs = sym.diff(fx*gx)
rhs = sym.diff(fx)*gx + sym.diff(gx)*fx

# confirming equality:
display(Math("(fg)^{'} = %s "%(sym.latex(lhs))))
display(Math("f^{'}g + g^{'}f = %s "%(sym.latex(lhs))))

# conclusion:
print("Yes! They are the same!!! ")
```

# exercise 2: confirm the quotient rule

# $$ \left(\frac{fx}{gx}\right)^{'} == \frac{f^{'}g - g^{'}f }{g^{2}} $$

```python
# exercise 2: confirm the quotient rule

x = sym.symbols('x')
fx = x**2
gx = sym.cos(x)

# prove: (f/g)^{'} = (f^{'}g - g^{'}f)/g^{2}
lhs = sym.diff(fx/gx)
rhs = (sym.diff(fx)*gx - sym.diff(gx)*fx)/(gx**2)

# confirming Equality:
display(Math("\\left(\\frac{fx}{gx}\\right)^{'} = %s "%(sym.latex(lhs))))
display(Math("\\frac{f^{'}g - g^{'}f }{g^{2}} = %s "%(sym.latex(rhs))))

# conclusion:
print("Yes! They are the same!!! ")
```

# Implicit Differentiation in sympy



## goal for problem 1:
1. use sym.idiff to compute derivative of xy = 1
2. use sym.plot_implicit() to plot the function and its derivative.
3. use sym.solve to transform the function to its explicit form.
4. substitute the explicit form in the implicit form to get y^{'} without y.
5. plot the derivative.

```python
x,y = sym.symbols('x,y')

# xy-1 = 0 
fx = x*y-1

# using sym.idiff, args are:(function,y,x)
implicit_diff = sym.idiff(fx,y,x)

print('\n implicit diff: ')
display(implicit_diff)

# plot the implicit diff
print("\n implicit differentiation plot: ")
sym.plot_implicit(fx);

# use sym.solve to transform the function to its explicit form
explicit_form = sym.solve(fx,y)[0]

print('\n Explicit form:')
display(explicit_form)

# substitute the explicit form in the implicit form to get y^{'} without y.
substituted_implicit = implicit_diff.subs(y,explicit_form)

print('\n solution: ')
display(substituted_implicit)

#  plot the derivative
print('\n Actual Derivative of the function plot: ')
p = sym.plot(substituted_implicit,show=False)
p.xlabel= 'X'
p.ylabel = 'Y'
p.ylim = [-50,5]
p.show()
```

* it is all about `measuring change and change of change(s)`.

# L'HOPITAL'S RULE: for indeterminant limits.

## $$ \lim_{x \rightarrow constant} f(x) = \frac{0}{0} \text{, then apply Lhopital's rule} $$
## $$ \lim_{x \rightarrow \infty} f(x) = \frac{\infty}{\infty} \text{, then apply Lhopital's rule} $$
## $$ \lim_{x \rightarrow \infty} f(x) = \frac{constant}{\infty} \text{, then answer is 0} $$


## Rule violation:
* **if the limit of the ratio of two function is indeterminate then you can try solving the problem by taking their derivatives**.

![fsfsfsd](Images/170124_calc2.png)

## notes
* ratio of the derivatives is not equal to the ratio of function of derivatives to which you'd apply quotient rule.


## formula for l'hopital:
# $$ \lim_{x \rightarrow 0} \frac{f(x)}{g(x) } = \frac{0}{0} \text{ then take } \rightarrow \lim_{x \rightarrow 0} \frac{df}{dx} $$

# $$ \lim_{x \rightarrow \infty} \frac{f(x)}{g(x) } = \frac{\infty}{\infty} \text{ then take } \rightarrow \lim_{x \rightarrow \infty} \frac{df}{dx} $$

![fsdfsfs](Images/180124_calc0.png)

### Notes:
* `if you don't get the answer and still get 0/0 or infinity/infinity in the first derivative then take second derivative or third derivative or fourth derivative till you get a constant as the answer keep on taking derivatives of the function.`

## application:

### Notes:
* if you don't get the answer and still get 0/0 or infinity/infinity in the first derivative then go for the second and third derivatives.

![sdfsfs](Images/180124_calc3.png)
### Notes:
* convert the functions to the form of f(x)/g(x) 
* when you get a definite value such as a constant stop taking further derivatives.


## conditions for L'hopital's rule:

## conflicting examples where L'hopitals can't be applied:
![sdfsdfsd](Images/180124_calc5.png)

## summary:
* apply L'hopitals rule only when the answer to the limit of the function is indeterminate with either 0/0 or infinity/infinity.
* apply L'hopitals rule only when the function is in the ratio form: f(x)/g(x) and gives indeterminate answers.
* take the second and third derivative only when the first derivative gives only indeterminate answer such as 0/0 or infinity/infinity.

```python
x = sym.symbols('x')
fx = x**2 * sym.log(x)

print("The limit of the function is: %s "%sym.limit(fx,x,0,'+-'))

# finding the minima of the function via second derivative
print("The function is: ")
display(fx)

second_der = sym.diff(sym.diff(fx,x),x)
print("The second derivative of the function: ")
display(second_der)

critical_points = sym.solve(second_der,x)[0]
print('The critical points are: ')
display(critical_points)

minima = second_der.subs({'x':critical_points})
print('The minima point is: ')
display(minima)

# plotting
p = sym.plot(fx,show=False)

p.show()
```

## notes:
* Rolles theorem states that if any two points of a function gives same output (dependent value) then there must exist atleast one point between those points whose derivative is exactly 0.
* in the picture above, the derivative of the torquise color point is exactly zero.

```python
def solveMVT(expr,a,c):
    from sympy.calculus.util import continuous_domain
    
    """
    Args:
    expr: expr you want to find the mean value theorem for
    a   : a point on the function to find b
    c   : a point on the function to find b
    
    
    returns: the value for b.
    the value for b is squashed between a and c given that the function is continuous between [a,c] 
    and differentiable between (a,c)
    """
    
    check = continuous_domain(expr,x,sym.Interval(a,c))
    if check:
        
        ## differentiate the function with b
        der_b = sym.diff(expr,x)
    
        ## solve for the equation
        b = sym.solve( der_b -  (expr.subs({'x':c})-expr.subs({'x':a}))/(c-a),x    )
        
        ## convert the sympy b to numpy b for computation purpose
        b = np.array(b).astype(np.float64)

        if len(b) > 0:
            # only return the b that is between a<b<c
            return b[np.bitwise_and(a<b,b<c)]

        else:
            return None


    return None

```

```python
def MakeThePlot(expr,a,c):
    
    def solveMVT(expr,a,c):
        from sympy.calculus.util import continuous_domain

        """
        Args:
        expr: expr you want to find the mean value theorem for
        a   : a point on the function to find b
        c   : a point on the function to find b


        returns: the value for b.
        the value for b is squashed between a and c given that the function is continuous between [a,c] 
        and differentiable between (a,c)
        """

        check = continuous_domain(expr,x,sym.Interval(a,c))
        if check:

            ## differentiate the function with b
            der_b = sym.diff(expr,x)

            ## solve for the equation
            b = sym.solve( der_b -  (expr.subs({'x':c})-expr.subs({'x':a}))/(c-a),x    )
        
            ## convert the sympy b to numpy b for computation purpose
            b = np.array(b).astype(np.float64)

            if len(b) > 0:
                # only return the b that is between a<b<c
                return b[np.bitwise_and(a<b,b<c)]
            else:
                return None
        return None


    # plotting
    xx = np.linspace(a-2,c+2)
    expr_numpy = sym.lambdify(x,expr,'numpy')

    plt.figure(dpi=150)
    plt.plot(xx,expr_numpy(xx),label='f(x)')

    # plotting point 'a' and the vertical dotted line
    plt.plot(a,expr_numpy(a),'ro',label='a')
#     plt.ylim([-5,25])
    axes = plt.gca()
    plt.plot([a,a],[axes.get_ylim()[0],axes.get_ylim()[1]],'r:',alpha=0.6)

    # plotting point 'c' and the vertical dotted line
    plt.plot(c,expr_numpy(c),'mo',label='c')
    plt.plot([c,c],[axes.get_ylim()[0],axes.get_ylim()[1]],'m:',alpha=0.6)

    # plottin the secant line(chord)
    plt.plot([a,c],[expr_numpy(a),expr_numpy(c)],'g--',alpha=0.4,label='secant')

    # plotting the point 'b' and the tangent line
    b = solveMVT(expr,a,c)
    for i,b_value in enumerate(b):
        plt.plot(b_value,expr_numpy(b_value),'ko',label='$b_{%s}$'%(i))
        plt.plot([b_value,b_value],[axes.get_ylim()[0],axes.get_ylim()[1]],'k:',alpha=0.3)
        tangent_x = np.array([b_value-1,b_value+1]) # just randomly chosen points
        tangent_y = sym.diff(expr,x).subs(x,b_value)*(tangent_x-b_value) + expr.subs(x,b_value)
        plt.plot(tangent_x,tangent_y,'k--',alpha=0.7,label='$tangent_{%s}$'%(i))


    plt.legend(loc=(1.1,0.5))
    plt.show()
```

```python
# exercise 1: part 2

def PlotFun(expr):
    ## lambdify the expr
    xx = np.linspace(-8,8,1000)
    expr_numpy = sym.lambdify(x,expr,'numpy')
    expr_numpy = expr_numpy(xx)

    ## compute empirical derivative
    dx = np.diff(xx)[0]
    df_dx_emp = np.diff(expr_numpy,1)/dx

    ## plot both the function and the empirical derivative 
    fig,ax = plt.subplots(1,2,figsize=(10,5))

    ax[0].plot(xx,expr_numpy)
    ax[0].set_title('$ fx = %s $'%(sym.latex(expr)))
    ax[0].set_xlabel('X')
    ax[0].set_ylabel('Y')

    ax[1].plot(df_dx_emp)
    ax[1].set_title('$ \\frac{df}{dx} $')
    ax[1].set_xlabel('X')
    ax[1].set_ylabel('Y')
    # ax[1].set_xlim([-10,10])


    plt.show()

    ## determine appropriate domain from sympy
    from sympy.calculus.util import continuous_domain
    from sympy import oo

    print("\n Domain for the function is: ")
    display(continuous_domain(expr,x,sym.Interval(-oo,oo)))

    ## sympy computed derivative
    print("\n Here's the answer for derivative of the expression: \n")
    display(sym.diff(expr,x))

    p = sym.plot(sym.diff(expr),show=False)
    # p.ylim = [-6,6]

    p.show()
    
PlotFun(expr)
```
