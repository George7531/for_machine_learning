# Linear and NonLinear Equations

# Linear:

1. A function is Linear if it produces a line.


2. A Linear function can be simplified with a equation: y = mx + c


3. diff in y / diff in x = constant, if it is linear. i.e y = 2x = 2 is a linear function.



4. Linear vs affine: Linear lines pass through the origin(0,0) where as affine lines do not but may totally resemble a linear line except that it does not pass through the origin. 

# NonLinear:

1. A function is NonLinear if it doesn't produce a Line.

2. A Non Linear function cannot be simplified with a equation: y = mx + c

3. diff in y / diff in x  constant, if it is linear. i.e y = x^2 =  2x + c is a non linear function.

```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
sym.init_printing()
import matplotlib.pyplot as plt
```

```python
# setting the degree threshold
degree = 3

# setting the random_coefficients
random_coeff = np.random.randn(degree+1)

# defining the x symbol to be included inside the eq
x = sym.symbols('x')

# eq = 0 because we are going to iterate and increment
eq = 0

# iteration and equation components aggregation is going on in here.
for i in range(len(random_coeff)):
    eq = eq + random_coeff[i]*x**(i)

# converting the sympified eq into the equivalent numpy version of it using lambdify function
numpy_func = sym.lambdify(x,eq,'numpy')

# x replacement
xx = np.linspace(-5,5,101)

# plotting and labeling
plt.plot(xx,numpy_func(xx))
plt.xlabel('X',fontsize=15)
plt.ylabel('Y=f(x)',fontsize=15)
plt.title('$ Y = %s $'%sym.latex(eq))

# showing
plt.show()
```

# Maclaurin Series

# $$ sin(x) = \sum_{n=1}^{\infty} (-1)^{n+1} \frac{x^{2n-1}}{(2n-1)!} $$

```python
# we are replacing the infinity with the fixed end for simplification.
end = 10


# defining the x:
x = np.linspace(-2*np.pi,2*np.pi,100)


# defining the function:
sin = 0

# for summation
for n in range(1,end+1):
    
    sin = sin + (-1)**(n+1) * (x**(2*n-1))/np.math.factorial(2*n - 1)
    
    # plotting all the lines
    plt.plot(x,sin,linestyle='dashed')

# plotting the final sine wave in black
plt.plot(x,sin,'k',linewidth=5,label='maclaurin series sin wave')

# plotting the origin lines
axes = plt.gca()

plt.plot(axes.get_xlim(),[0,0],'k--')
plt.plot([0,0],axes.get_ylim(),'k--')

# setting the limit on y and x for zoomin in
plt.ylim(-4,4)
plt.xlim(x[0],x[-1])


# verifying the authenticity of maclaurin series being able to produce real sine wave
plt.plot(x[::4],np.sin(x[::4]),'bo',markerfacecolor='white',linewidth=2,label = 'sin(x)')

# showing legend
plt.legend(loc='lower left')


# showing the work
plt.show()


```

# piece wise functions

```python
# exercise 1

x = np.linspace(-5,5,1000)
y = np.zeros(len(x))

for idx,val in enumerate(x):
    
    # piece 1
    if val < 0:
        y[idx] = 0
    
    # piece 2
    elif (val >= 0 and val < 3):
        y[idx] = -2*val
    
    # piece 3
    elif val >= 3:
        y[idx] = 0.1*val**3
        
plt.figure(dpi=150)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y=f(x)')


plt.show()
#  this is O(N) solution
```

```python
# more efficient solution: this is just O(1)

# exercise 1.
x = np.linspace(-3,5,100)

# constructing piece of functions:
piece1 = (0)        * (x<0)  # value * condition
piece2 = (-2*x)     * ((x>=0) & (x<3))
piece3 = (0.1*x**3) * (x>=3)

# stiching all the pieces together
y = piece1+piece2+piece3


plt.plot(x,y)

plt.show()
```

```python
# exercise 2:

x = np.linspace(-3,5,100)

# when x<0
indices1 = np.where(x < 0)
piece1   = x[indices1] * 0

# when 0 <= x < 3
indices2 = np.where((x >= 0) & (x < 3))
piece2   = -2*x[indices2]

# when x>=3
indices3 = np.where(x >= 3)
piece3   = 0.1*x[indices3]**3


# plotting
plt.plot(x[indices1],piece1)
plt.plot(x[indices2],piece2)
plt.plot(x[indices3],piece3)

plt.show()
```

# continuous and discontinuous functions

# continuous functions:

* they have the values in between. they do not have any missing values between any point on the function

# Discontinuous functions:


* they got missing values in between


# jump discontinuity:
* values are missing in certain areas and the function seem to jump from one space to another without any in-between values filling in some space

# infinite discontinuity:
* function tend towards infinity between two places where the function has become discontinuous

# removable discontinuity:
* function is discontinous only in one place due to it going to infinity at one point in space



# A function is continuous if these three conditions are met: 

# 1. $$ \text{ f(a) is defined } $$
# 2. $$ \lim_{x \rightarrow a} f(x) \text{exists} $$

$$ f(x) = x^3 $$

$$ f(-x) = (-x)^3 = -(x^3) = -f(x) $$

# symmetric and asymmetric:

* symmetricity means any side is as same as the other. assymmetricity means disproportionate,disfigured, unhaveled, unlevelled, imbalanced. 

* functions that obey the rules of even or odd are symmetric and functions that don't are assymetric.

* function is symmetric if f(-x) = f(x) or -f(x)

* functions can be even-symmetric,odd-symmetric and asymmetric.

# prove:

1. prove that angle of the multiplication of two complex numbers is equal to the **addition of angles** of both of them.

2. prove that length of the multiplication of two complex numbers is equal to the **multiplication of lengths** of both of them.

```python
# exercise 1

# 1.  (4+3j)j

# real for x and imag for y.

eq1 = 4+3j
eq2 = 1j
multiply = eq1*eq2

# plotting the equations
plt.plot([0,eq1.real],[0,eq1.imag],label='$ %s $'%(sym.latex(sym.sympify(eq1))))
plt.plot([0,eq2.real],[0,eq2.imag],label='$ %s $'%(sym.latex(sym.sympify(eq2))))
plt.plot([0,multiply.real],[0,multiply.imag],label='$ %s $'%(sym.latex(sym.sympify(multiply))))

# setting the xlims and ylims
plt.xlim([-5,5])
plt.ylim([-5,5])

# setting the origins
axes = plt.gca()
plt.plot(axes.get_xlim(),[0,0],'k--',alpha=0.3)
plt.plot([0,0],axes.get_ylim(),'k--',alpha=0.3)

# setting the labels for x axis and y axis
plt.xlabel('Negative and Positive')
plt.ylabel('Imaginary')

plt.legend(loc=(1.1,0.5))

plt.show()


# printing angles and lengths of equations

def angle(eq):
    # eq.real == 0 arctan(x/0) = infinity == 90 degree
    if eq.real == 0:
        return 90
    
    a = np.rad2deg(np.arctan(eq.imag/eq.real)) # as in arctan(y/x)
    
    # for negative angles add 180 degree.
    if a < 0:
        return a+180
    
    return a
    


angle1 = angle(eq1) 
angle2 = angle(eq2)
angle_multiplied = angle(multiply)  

def length(eq):
    return np.linalg.norm(eq)

length1 = length(eq1)
length2 = length(eq2)
length3 = length(multiply)

print('Angle for eq1: {}, length for eq1: {}'.format(angle1,length1))
print('Angle for eq2: {}, length for eq2: {}'.format(angle2,length2))
print('Angle for their multiplication: {}, length for multiplied: {}'.format(angle_multiplied,length3))
```

```python
# exercise1:

# creating random polynomial function of this format: y = constant + coeff1*x + coeff2 * x**2


def random_polynomial(n):
    # range of random coefficients:
    xx = np.random.randint(-5,5,n)


    # building the function

    # initiating the function
    fx = str(xx[0])

    # getting into the loop
    for i in range(1,n):

        # if xx[i] is a negative number use '-' instead of '+'
        if xx[i] < 0:

            # if the power is 1 use just 'x' instead of 'x**1' that would look pristine!
            # we don't have to worry about zero since zero is out of equation.
            if i == 1:


                # if the coefficient is -1. we want it to be printed as -x not as -1x right? so we write the code for it.
                if xx[i] == -1:
                    fx = fx + '-x'

                # if the coefficient is other than -1    
                else:
                    fx = fx  + str(xx[i]) + 'x'

            # if the power is anything other than 1
            else:
                # if the coefficient is -1
                if xx[i] == -1:
                    fx = fx + '-x**' + str(i)

                # if the coefficient is other than -1
                else:
                    fx = fx  + str(xx[i]) + 'x**' + str(i)



        # if xx[i] is a positive number. we use '+'
        else:
            # if the power is 1
            if i == 1:


                # if the coefficient is 1
                if xx[i] == 1:
                    fx = fx + '+' + 'x'

                # if the coefficient is other than 1
                else:
                    fx = fx  + "+" + str(xx[i]) + 'x'

            # if the power is something else other than 1
            else:
                # if the coefficient is 1
                if xx[i] == 1:
                    fx = fx + '+' + 'x**' + str(i)

                # if the coefficient is other than 1
                else:
                    fx = fx  + "+" + str(xx[i]) + 'x**' + str(i)
                    
    return [fx,xx]



        
        
            
# I know this is bad coding but we could improve the code in time don't feel guilty now! (a message to future self(ves))
```

```python
import numpy as np
import sympy as sym
from IPython.display import Math,display
sym.init_printing()
import matplotlib.pyplot as plt
```

```python
# exercise 3: random piece-wise function

x = sym.symbols('x')

# choose one of the functions for different pieces.
part1 = [x,x**2,x**3] # for x<=0
part2 = [sym.log(x),sym.sqrt(x)] # for x>0

xx = np.linspace(-3,3,100)

# for x<=0
index1 = np.where(xx<=0)
func1 = np.random.choice(part1)
piece1_func = sym.lambdify(x,func1,'numpy')


# for x>0
index2 = np.where(xx>0)
func2 = np.random.choice(part2)
piece2_func = sym.lambdify(x,func2,'numpy')


# printing the functions to sketch:
print('Sketch this!')
display(Math('%s  \\quad \\text{ for x<=0 }'%(sym.latex(func1))))
display(Math('%s \\quad \\text{ for x>0 }'%(sym.latex(func2))))

# plotting the pieces
plt.figure(dpi=100)
plt.plot(xx[index1],piece1_func(xx[index1]),label='$ %s $'%(sym.latex(func1)))
plt.plot(xx[index2],piece2_func(xx[index2]),label='$ %s $'%(sym.latex(func2)))

plt.legend()
plt.show()
```
