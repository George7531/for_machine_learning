```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# this is the range
xx = np.linspace(-3,3,51)

# the grid has many repition of the range we mentioned above forming a grid like matrix. 
X,Y = np.meshgrid(xx,xx)

# random values

f1 = 4
f2 = 0.8 

# it is a function of X and Y
Z = np.sin(f1*X,np.cos(f2*Y))

# taking the derivative of z
gx,gy = np.gradient(Z)


# plotting 

plt.figure(dpi=200)

plt.contourf(xx,xx,Z,40)

plt.quiver(xx,xx,gx,gy)

plt.colorbar()
plt.show()
```
