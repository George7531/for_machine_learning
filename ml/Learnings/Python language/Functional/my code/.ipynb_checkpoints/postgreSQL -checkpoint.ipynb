```python
import psycopg2 as pg2
```

```python
# establish the connection with postgresql;
conn = pg2.connect(database='dvdrental',user='postgres',password='loop')
```

```python
# create a cursor for the connection to move around the database.
cur = conn.cursor()
```

```python
# move the cursor to a specific location in the database (run this everytime you want to fetch the data)
cur.execute('select * from payment')
```

```python
# fetch all the data from the payment column
cur.fetchall()
```

```python
# you can also select specified number of rows
data = cur.fetchmany(5)
```

```python
data
```

```python
# manipulate the data like a list
data[0]
```

```python
# close the connection
conn.close()
```
